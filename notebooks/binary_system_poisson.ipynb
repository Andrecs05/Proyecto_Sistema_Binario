# Gravitational potential and field of a binary star system

Numerical solution to Poisson's equation

```python
%matplotlib widget
import sys
from pathlib import Path

sys.path.insert(0, str(Path.cwd().parent))

import numpy as np
import matplotlib.pyplot as plt
from scripts.plotters import *
from scripts.solvers import *
```

```python
M1 = 2.0
M2 = 1.0
a = 1.0
G = 1.0
R_star1 = 0.2
R_star2 = 0.1
L = 1.2
grid_size = 100

x = np.linspace(-L, L, grid_size)
y = np.linspace(-L, L, grid_size)
X, Y = np.meshgrid(x, y)

phi = binary_potential_poisson_fd(M1, M2, R_star1, R_star2, a, G, L, grid_size)

plot_potential_contours(phi, M1, M2, a, size=L, res=grid_size, levels=100, point_masses=False, R_star1=R_star1, R_star2=R_star2)
plot_potential_transverse(phi, x, y, y0=0, point_masses=False, M1=M1, M2=M2, R_star1=R_star1, R_star2=R_star2, a=a)
plot_binary_potential_3D(phi, X, Y)


g_x, g_y = binary_field_laplace(phi, X, Y)
plot_field_streamlines(g_x, g_y, X, Y, M1, M2, a, density=1.5, point_masses=False, R_star1=R_star1, R_star2=R_star2)
```
